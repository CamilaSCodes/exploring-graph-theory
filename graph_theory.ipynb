## Graph Shortest Path

```python
edges=[('A','B',1),('A','F',8),('A','E',4),('B','C',2),('B','G',6),('B','F',6),('C','D',1),('C','G',2),('D','G',1),('D','H',4),('E','F',5),('G','F',1),('G','H',1)]

Z = DiGraph(edges, weighted=True)

Z.plot(edge_labels=True)
```

```python
from sage.graphs.base.boost_graph import shortest_paths

print("\nShortest path from 'A' using Dijkstra's Algorithm:",shortest_paths(Z, 'A', algorithm='Dijkstra'))
```

```python
edges2=[('A','B',1),('A','E',4),('B','C',2),('C','D',1),('C','G',2),('G','F',1),('G','H',1)]

Y = DiGraph(edges2, weighted=True)

Y.plot(edge_labels=True)
```

```python
print("\nShortest path from 'A' using Bellman-Ford Algorithm:",shortest_paths(Y, 'A', algorithm='Bellman-Ford_Boost'))
```

```python
X = DiGraph([('A','B',{'weight':4}),('A','C',{'weight':-2}),('B','G',{'weight':-2}),('B','H',{'weight':-4}),('C','D',{'weight':2}),('C','F',{'weight':1}),('G','I',{'weight':-1}),('I','H',{'weight':1}),('H','G',{'weight':1}),('F','D',{'weight':3}),('E','H',{'weight':3}),('S','A',{'weight':7}),('S','C',{'weight':6}),('S','F',{'weight':5}),('S','E',{'weight':6}),('E','F',{'weight':-2})])

X.plot()
```

```python
X.shortest_path_lengths('A', algorithm='Bellman-Ford_Boost')
print("\nLenght of the shortest path from 'A' to any other vertex using Bellman-Ford Algorithm:",X.shortest_path_lengths('A', algorithm='Bellman-Ford_Boost'))
```
